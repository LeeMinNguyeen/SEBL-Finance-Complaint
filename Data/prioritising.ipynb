```python
import pandas as pd
```

```python
# Set Risk variables
# 1
Operation_Risk = [
    "Credit Reporting & Identity Protection",
    "Account Management Issues",
    "Digital Wallet & Online Services",
    "Other or Unclassified Issues",
    "Customer Asset Repossession & Damage",
    "Payment & Transaction Errors"
]

# 3
Financial_Risk = [
    "Credit & Payment Card Issues",
    "Loan & Mortgage Problems"
]

# 4
Strategic_Risk = [
    "Misleading Information & Advertising",
    "Poor Service & Broken Promises",
    "Unexpected Fees & Charges"
]

# 2
External_Risk = [
    "Fraud, Scams & Threatening Behavior",
    "Data Breaches & Cybersecurity Issues",
    "Regulatory & Compliance Concerns"
]
```

```python
# Set Priority variables
# 3
Highly_Likely = [
    "Credit Reporting & Identity Protection",
    "Account Management Issues",
    "Credit & Payment Card Issues",
    "Poor Service & Broken Promises"
]

# 2
Likely = [
    "Fraud, Scams & Threatening Behavior",
    "Other or Unclassified Issues",
    "Unexpected Fees & Charges",
    "Payment & Transaction Errors",
    "Loan & Mortgage Problems"
]

# 1
Unlikely = [
    "Misleading Information & Advertising",
    "Asset Repossession & Damage",
    "Credit Reporting & Identity Protection",
    "Digital Wallet & Online Services"
]
```

```python
df = pd.read_csv("Complaint_Dataset_Unlabel.csv")
df
```

```python
for idx, row in df.iterrows():
    if row['Issue_Group'] in Operation_Risk:
        df.at[idx, 'Risk'] = 1
    elif row['Issue_Group'] in Financial_Risk:
        df.at[idx, 'Risk'] = 3
    elif row['Issue_Group'] in Strategic_Risk:
        df.at[idx, 'Risk'] = 4
    else:
        df.at[idx, 'Risk'] = 2
    
    if row['Issue_Group'] in Highly_Likely:
        df.at[idx, 'Priority'] = 3
    elif row['Issue_Group'] in Likely:
        df.at[idx, 'Priority'] = 2
    else:
        df.at[idx, 'Priority'] = 1

df['Priority_Score'] = df['Priority'] * df['Risk']
df['Priority_Rating'] = df['Priority_Score'].apply(
    lambda x: 'Critical' if x >= 6 else ('Medium' if 3 <= x < 6 else 'Low')
)
```
